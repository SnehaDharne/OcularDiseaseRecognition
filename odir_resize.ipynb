```python
from PIL import Image
import os

input_dir = data_dir_cropped
output_dir = data_dir_resized

# Set the desired image size
size = (256, 256)

# Loop through all images in the input directory
for filename in os.listdir(input_dir):
    # Open the image
    img = Image.open(os.path.join(input_dir, filename))

    # Resize the image
    img = img.resize(size)

    # Save the resized image to the output directory
    img.save(os.path.join(output_dir, filename))
```

```python
import os

src_files = os.listdir(data_dir_cropped)
dst_files = os.listdir(data_dir_resized)

missing_files = []
for file in src_files:
    if file not in dst_files:
        missing_files.append(file)

if missing_files:
    print("The following files are missing in the destination folder:")
    print(missing_files)
else:
    print("All files are present in the destination folder.")
```
