```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import TomekLinks
from sklearn.preprocessing import OneHotEncoder

# Load the dataset into a pandas DataFrame
data = pd.read_csv(path)

# Separate the features (image IDs) and labels
X = data['ID']
y = data[['Normal', 'Diabetes', 'Glaucoma', 'Cataract', 'AMD', 'Hypertension', 'Myopia', 'Others']]

# Create a new DataFrame for the balanced dataset
balanced_data = pd.DataFrame(columns=['ID', 'Normal', 'Diabetes', 'Glaucoma', 'Cataract', 'AMD', 'Hypertension', 'Myopia', 'Others'])

# Encode the "ID" column using OneHotEncoder
encoder = OneHotEncoder(sparse=False)
X_encoded = encoder.fit_transform(X.values.reshape(-1, 1))

# Balance each label separately
for column in y.columns:
    label = y[column]

    # Apply SMOTE oversampling
    smote = SMOTE()
    X_oversampled, y_oversampled = smote.fit_resample(X_encoded, label)

    # Apply TomekLinks undersampling
    tomek = TomekLinks()
    X_resampled, y_resampled = tomek.fit_resample(X_oversampled, y_oversampled)

    # Convert the encoded features back to the original string identifiers
    X_resampled_decoded = encoder.inverse_transform(X_resampled)

    # Create a temporary DataFrame for the balanced label
    balanced_label = pd.DataFrame({'ID': X_resampled_decoded.flatten(), column: y_resampled})

    # Concatenate the temporary DataFrame to the balanced dataset
    balanced_data = pd.concat([balanced_data, balanced_label], ignore_index=True)

# Save the balanced dataset to a CSV file
# balanced_data.to_csv('balanced_dataset.csv', index=False)
```

```python

# Iterate over the rows of the balanced dataset
for index, row in balanced_data.iterrows():
    id_value = row['ID']

    # Find the corresponding row in the previous dataset
    previous_row = df[df['ID'] == id_value]

    # Check if a matching row is found
    if len(previous_row) > 0:
        # Fill null values in the balanced dataset with corresponding values from the previous dataset
        balanced_data.loc[index] = balanced_data.loc[index].fillna(previous_row.iloc[0])
    else:
        print(f"No matching row found for ID: {id_value}")

```

```python
import pandas as pd

# Load the dataset into a pandas DataFrame
data = balanced_data

# Calculate the desired occurrence for each ID
desired_occurrence = 8

# Group the data by ID and calculate the current occurrence for each ID
current_occurrence = data['ID'].value_counts()

# Initialize an empty DataFrame for the filtered dataset
filtered_data = pd.DataFrame()

# Iterate over each unique ID
for ID, group in data.groupby('ID'):
    # Calculate the number of samples to keep for this ID
    samples_to_keep = max(current_occurrence[ID] // desired_occurrence, 1)

    # Sample the data to keep the desired number of samples
    subset = group.sample(n=samples_to_keep, random_state=42)

    # Append the subset to the filtered dataset
    filtered_data = pd.concat([filtered_data, subset])

# Reset the index of the filtered dataset
filtered_data = filtered_data.reset_index(drop=True)

```
